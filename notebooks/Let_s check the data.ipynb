# Processing dates

```python
# converting date times
cols = ['Date.of.Birth', 'DisbursalDate']
a = []
for col in cols:
    df[col] = df[col].apply(lambda x: dt.datetime.strptime(x, '%d-%m-%y'))
df.head()
```

```python
regex = re.compile(r'((?P<year>\d+?)yrs)?\s((?P<month>\d+?)mon)')
def parseDate(date_str):
    parts = regex.match(date_str)
    if not parts:
        return
    parts = parts.groupdict()
    days = 0
    for k, v in zip(parts.keys(), parts.values()):
        if v:
            days += int(v)*360 if k == 'year' else int(v)*30
    return days
```

```python
cols = ['AVERAGE.ACCT.AGE', 'CREDIT.HISTORY.LENGTH']
for col in cols:
    df[col] = df[col].apply(lambda x: parseDate(x))
df.head()
```

```python
df['AgeWhenPaied'] = (df['DisbursalDate'] - df['Date.of.Birth']).dt.total_seconds()/(60*60*24*360)
df.head()
```

```python
# people with negative age who are doing loan default
len(df[np.logical_and(df['AgeWhenPaied'] <= 0, df['loan_default'] == 1)])
```

```python
# people with negative age
df['FalseAge'] = df['AgeWhenPaied'] <= 0
df['FalseAge'] = list(map({True:1, False:0}.get, df['FalseAge']))
df.head()
```

# Encoding labels

```python
from sklearn.preprocessing import LabelEncoder
le_for_empType = LabelEncoder()
le_for_description = LabelEncoder()

le_for_empType.fit(list(df['Employment.Type']))
le_for_description.fit(list(df['PERFORM_CNS.SCORE.DESCRIPTION']))

df['Employment.Type'] = le_for_empType.transform(list(df['Employment.Type']))
df['PERFORM_CNS.SCORE.DESCRIPTION'] = le_for_description.transform(list(df['PERFORM_CNS.SCORE.DESCRIPTION']))
```

```python
df = pd.read_csv(train_path + 'new_train.csv')
df.head()
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, f1_score, roc_auc_score
import gc
from sklearn.naive_bayes import GaussianNB 
```

```python
# columns to train the model on 
colsToTrain = [col for col in df.columns if col not in ['UniqueID', 'loan_default', 'Date.of.Birth', 'DisbursalDate']]
```

```python
y = df['loan_default']
df.drop(inplace=True, axis=1, columns='loan_default')
x = df
del df
gc.collect()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=2019, test_size=0.3)     
```

```python
model = XGBClassifier()
model.fit(x_train[colsToTrain], y_train)
pred = model.predict(x_test[colsToTrain])
```

```python
print(f'f1 score of the model is {f1_score(y_test, pred)}')
print('following is the confusion matrix:')
print(confusion_matrix(y_test, pred))
print(f'area under the curve score is {roc_auc_score(y_test, pred)}')
```

```python
NBmodel = GaussianNB()
NBmodel.fit(x_train[colsToTrain], y_train)
NBpred = NBmodel.predict(x_test[colsToTrain])
```

```python
print(f'f1 score of the model is {f1_score(y_test, NBpred)}')
print('following is the confusion matrix:')
print(confusion_matrix(y_test, NBpred))
print(f'area under the curve score is {roc_auc_score(y_test, NBpred)}')
```
